```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
train_host_info = pd.read_csv('train_host_info.csv')
train_iot_info = pd.read_csv('train_iot_info.csv')
train_matrl_id_info = pd.read_csv('train_matrl_id_info.csv')

def read_year(x,length):
    return round(x/length)
def read_month(x):
    return int(str(x)[4:6])
train_matrl_id_info['year'] = train_matrl_id_info['sale_time'].map(lambda x: read_year(x,10000))
train_matrl_id_info['month'] = train_matrl_id_info['sale_time'].apply(read_month)
train_host_info['year'] = train_host_info['month'].map(lambda x: read_year(x,100))
train_host_info['month_'] = train_host_info['month'].apply(read_month)
train_iot_info['year'] = train_iot_info['week'].map(lambda x: read_year(x,10000))
train_iot_info['month'] = train_iot_info['week'].apply(read_month)
```

```python
typeset = sorted(list(set(train_host_info['type'])))
train_host_info_month = pd.DataFrame(columns=["matrl_id","year","month"] + [f'host_nums_{tp}' for tp in typeset])
for i in tqdm(range(1,1201)):
    for year in range(2018,2021):
        if year < 2020:
            for m in range(1,13):
                month_host = train_host_info[(train_host_info['matrl_id']==i) & (train_host_info['year']==year) & (train_host_info['month_']==m)]
                host_dic = {"matrl_id":[i],"year": [year],"month":[m]}
                for tp in typeset:
                    tp_host = month_host[month_host['type']==tp]
                    if tp_host.shape[0] == 0:
                        host_dic[f'host_nums_{tp}'] = [0]
                    else:
                        host_dic[f'host_nums_{tp}'] = tp_host['host_nums'].values
                    
                temp = pd.DataFrame(host_dic)
                train_host_info_month=train_host_info_month.append(temp,ignore_index=True)
                
        else:
            for m in range(1,7):
                month_host = train_host_info[(train_host_info['matrl_id']==i) & (train_host_info['year']==year) & (train_host_info['month_']==m)]
                host_dic = {"matrl_id":[i],"year": [year],"month":[m]}
                for tp in typeset:
                    tp_host = month_host[month_host['type']==tp]
                    if tp_host.shape[0] == 0:
                        host_dic[f'host_nums_{tp}'] = [0]
                    else:
                        host_dic[f'host_nums_{tp}'] = tp_host['host_nums'].values
                    
                temp = pd.DataFrame(host_dic)
                train_host_info_month=train_host_info_month.append(temp,ignore_index=True)
```

```python
typeset = sorted(list(set(train_iot_info['type'])))
train_iot_info_month = pd.DataFrame(columns=["matrl_id","year","month"] + [f'work_time_{tp}' for tp in typeset] + [f'work_nums_{tp}' for tp in typeset])
for i in tqdm(range(1,1201)):
    for year in range(2018,2021):
        if year < 2020:
            for m in range(1,13):
                month_host = train_iot_info[(train_iot_info['matrl_id']==i) & (train_iot_info['year']==year) & (train_iot_info['month']==m)]
                # print(month_host)
                host_dic = {"matrl_id":[i],"year": [year],"month":[m]}
                for tp in typeset:
                    tp_host = month_host[month_host['type']==tp]
                    #print(tp_host)
                    if tp_host.shape[0] == 0:
                        host_dic[f'work_time_{tp}'] = [0]
                        host_dic[f'work_nums_{tp}'] = [0]
                    else:
                        host_dic[f'work_time_{tp}'] = sum(tp_host['work_time'].values)
                        host_dic[f'work_nums_{tp}'] = sum(tp_host['work_nums'].values)
                #print(host_dic)
                temp = pd.DataFrame(host_dic)
                #print(temp)
                train_iot_info_month=train_iot_info_month.append(temp,ignore_index=True)
                
        else:
            for m in range(1,7):
                month_host = train_iot_info[(train_iot_info['matrl_id']==i) & (train_iot_info['year']==year) & (train_iot_info['month']==m)]
                host_dic = {"matrl_id":[i],"year": [year],"month":[m]}
                for tp in typeset:
                    tp_host = month_host[month_host['type']==tp]
                    if tp_host.shape[0] == 0:
                        host_dic[f'work_time_{tp}'] = [0]
                        host_dic[f'work_nums_{tp}'] = [0]
                    else:
                        host_dic[f'work_time_{tp}'] = sum(tp_host['work_time'].values)
                        host_dic[f'work_nums_{tp}'] = sum(tp_host['work_nums'].values)
                    
                temp = pd.DataFrame(host_dic)
                train_iot_info_month=train_iot_info_month.append(temp,ignore_index=True)
```

```python
# 统计每个月的销量
train_matrl_id_sale = pd.DataFrame(columns=["matrl_id","year","month","month_sale"])
for i in tqdm(range(1,1201)):
    for year in range(2018,2021):
        if year < 2020:
            for m in range(1,13):
                month_sale = train_matrl_id_info[(train_matrl_id_info['matrl_id']==i) & (train_matrl_id_info['year']==year) & (train_matrl_id_info['month']==m)]
                temp = pd.DataFrame({"matrl_id":[i],"year": [year],"month":[m],"month_sale":[sum(month_sale['sale_nums'])]})
                train_matrl_id_sale=train_matrl_id_sale.append(temp,ignore_index=True)
        else:
            for m in range(1,7):
                month_sale = train_matrl_id_info[(train_matrl_id_info['matrl_id']==i) & (train_matrl_id_info['year']==year) & (train_matrl_id_info['month']==m)]
                temp = pd.DataFrame({"matrl_id":[i],"year": [year],"month":[m],"month_sale":[sum(month_sale['sale_nums'])]})
                train_matrl_id_sale=train_matrl_id_sale.append(temp,ignore_index=True)
            
```

```python
fea_col = sorted(list(set(list(main_df.columns)) - set(['matrl_id','year'])))
```

```python
import gc
def series_to_superviesed(x_timeseries, y_timeseries, n_memory_step, n_forcast_step, split = None):
    input_step, input_feature = x_timeseries.shape
    output_step, output_feature = y_timeseries.shape

    n_RNN_sample=input_step-n_forcast_step-n_memory_step+1
    RNN_x=np.zeros((n_RNN_sample,n_memory_step, input_feature))
    RNN_y=np.zeros((n_RNN_sample,n_forcast_step, output_feature))

    for n in range(n_RNN_sample):
        RNN_x[n,:,:]=x_timeseries[n:n+n_memory_step,:]
        RNN_y[n,:,:]=y_timeseries[n+n_memory_step:n+n_memory_step+n_forcast_step,:]
    if split != None:
        assert (split <=0.9) & (split >= 0.1), 'split not in reasonable range'
        return RNN_x[:int(split*len(RNN_x))], RNN_y[:int(split*len(RNN_x))],\
               RNN_x[int(split*len(RNN_x))+1:], RNN_y[int(split*len(RNN_x))+1:]
    else:
        return RNN_x, RNN_y, None, None

X, Y = [], []
for i in range(1,1201):
    df_i = main_df[main_df['matrl_id']==i].copy()
    x_timeseries = df_i[fea_col].values
    y_timeseries = df_i["month_sale"].values.reshape(-1,1)
    x_train,y_train,x_test,y_test = series_to_superviesed(x_timeseries,y_timeseries,12,3)
    if i == 1:
        print(x_train.shape)
        print(y_train.shape)
    X.append(x_train)
    Y.append(y_train)
    del x_train,y_train,x_test,y_test,df_i
    gc.collect()
```

```python
X = np.vstack(X)
Y = np.vstack(Y)
# np.save(f'X.npy',X)
# np.save(f'Y.npy',Y)
print(X.shape)
print(Y.shape)
```
